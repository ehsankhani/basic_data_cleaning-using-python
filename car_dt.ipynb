## import necessary libs
-**pandas** to handle the dataset(csv)

-**numpy** for the math

-**chardet** to detect the encodig of the dataset

```python
import numpy as np
import pandas as pd
import chardet
```

Detect the encoding

```python
with open('car_ad.csv', 'rb') as f:
    result = chardet.detect(f.read())

encoding = result['encoding']
print(f"Detected encoding: {encoding}")
```

Read the file with the detected encoding

```python
data = pd.read_csv('car_ad.csv', encoding=encoding)
data.head()
```

```python
#getting the shape of the raw dataset
data.shape
```

```python
data.info()  # Get information about the dataset (data types, non-null counts)
```

```python
data.describe()  # Get basic statistical summary
```

**first part: delete the missing values**

```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
data = data.dropna()  # Drop rows with any missing values
```

```python
data.shape  #getting the data shape to ensure that the missing values are gone
```

**second part: getting the duplicate counts**

```python
duplicates = data.drop_duplicates().sum()
print (duplicates.count())
```

```python
data = data.drop_duplicates()  # Drop duplicate rows
```

```python
data.shape  #getting the data shape to ensure that the duplicates are gone
```

replace the spacing and put ( _ )instead for the data columns

```python
data.columns = [col.strip().lower().replace(' ', '_') for col in data.columns]
data.head()
```
